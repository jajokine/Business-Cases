<h1 align='center'> A/B Testing </h1>

<b> What is A/B testing? </b>

A/B testing is an experiment on two variants to see which performs better based on a given metric. This helps to get a clear idea of what users prefer as we directly test one thing over the other, to determine what works better between the two options. Typically, the test is performed in a way where two consumer groups are exposed to two different versions of the same thing to see if there is a significant difference in metrics like average revenue, click-through rate, or bought products.

We run the experiment by randomly spliting our customer base into two groups, a control group and a treatment group. Then, we expose our treatment group with a new feature and see if we get a significant increase in the metric that we want to measure. This is done with a two-sample hypothesis testing where we determine whether the differences that we analyze through the metric between the two groups are statistically significant or not.

<b> Why is it important to know? </b>

For a business, this sort of testing gives valuable information when something new is being implemented and we want to know whether this implementation is worthwhile or we should stick to the old version or perhaps try something else. This could be, for example, when we try to understand user engagement and user satisfaction of online features such as a new feature or product. This analysis that utilizes statistical methods of hypothese testing allows companies to better understand growth, increase revenue, and optimize customer satisfaction. They are therefore an indispensable tool for many large companies, but also for smaller companies and startups by combining with Agile software development or through Minimum Viable Products (MVPs).

## How to conduct a A/B test 

<b> 1. Formulate the hypothesis </b>

Before conducting an A/B testing we want to state the null hypothesis and the alternative hypothesis:

The null hypothesis states that sample observations result purely from chance. From an A/B test perspective, the null hypothesis states that there is no difference between the control and treatment group.

The alternative hypothesis on the other hand, is one that states that sample observations are influenced by some non-random cause. From an A/B test perspective, the alternative hypothesis states that there is a difference between the control and treatment group.

When developing the null and alternative hypotheses we want to follow the PICOT format:

   - Population:     The group of people that participate in the experiment
   - Intervention:   The new feature or variant in the study
   - Comparison:     The reference group that is being compared against the intervention
   - Outcome:        The result we are measuring
   - Time:           The duration of the experience (when and how long the data is collected)

<b> 2. Create your control group and treatment group </b>

Once we determine the null and alternative hypothesis, the next step is to create the control and treatment group. Two important concepts that we need to consider in this step are random samplings and sample size.

<b>Random Sampling</b>

Random sampling is a technique where each sample in a population has an equal chance of being chosen. This is important because it eliminates sampling bias so that the results of the A/B test are representative of the entire population rather than the sample itself.

<b>Sample Size</b>

We need to determine the minimum sample size for the A/B test prior to conducting it so that we can eliminate under coverage bias that comes from sampling too few observations. 

<b> 3. Conduct the test, compare the results, and reject or do not reject the null hypothesis </b>

Once we collect the samples and conduct the experiment we need to determine if the difference between the control and the treatment group is statistically significant. There are a few steps in determining this:

   - First, we need to set a threshold for alpha, the probability of making a type 1 error. Typically the alpha is set at 5% or 0.05.
   
   - Next, we need to determine the probability value (p-value) by first calculating the test statistic using the right formula that depends on the assumed distribution of the business case. 
   
       - If we would like to model the average revenue per user, we would assume the samples come from a Normal distribution and we would perform a Welch t-test or a z-test depending on the sample sizes.
       - If we would like to model the click-through rate, we would assume the samples come from a Binomial distribution and we would perform a Fisher's exact test or Barnard's test if the sample size is small or a z-test with a large sample size.
       - If we would like to model the transactions per paying user, we would assume the samples come from a Poisson distribution and we would perform a Likelihood-ratio test.
       - If we would like to model the number of each product purchased, we would assueme the samples come from a Multinomial distribution and we would perform a Chi-Squared test.


   - Lastly, we compare the p-value to the alpha. If the p-value is greater than the alpha, we do not reject the null.

## 1. Designing the experiment

<b> Formulating a hypothesis </b>

First, we want to make sure we formulate a hypothesis at the start of our project. This will make sure our interpretation of the results is correct as well as rigorous.

Given we don’t know if the new design will perform better or worse as our current design, we’ll choose a two-tailed test:

    Hₒ: p = pₒ

    Hₐ: p ≠ pₒ

where p and pₒ stand for the conversion rate of the new and old design, respectively. We’ll also set a confidence level of 95%:

    α = 0.05

The α value is a threshold we set, by which we say “if the probability of observing a result as extreme or more (p-value) is lower than α, then we reject the Null hypothesis”. Since our α=0.05 (indicating 5% probability), our confidence (1 — α) is 95%.

All this really means is that whatever conversion rate we observe for our new design in our test, we want to be 95% confident it is statistically different from the conversion rate of our old design, before we decide to reject the Null hypothesis Hₒ.

<b> Creating the control and test groups </b>

After formulating the hypothesis we need to create two groups:

  -  Control group - To whom will be shown the old design
  -  Treatment (or experimental) group - They'll be shown the new design

This will be the <i>Independent Variable</i>. The reason we have two groups even though we know the baseline conversion rate is that we want to control for other variables that could have an effect on our results, such as seasonality: by having a control group we can directly compare their results to the treatment group, because the only systematic difference between the groups is the design of the product page, and we can therefore attribute any differences in results to the designs.

For the <i>Dependent Variable</i> (i.e. the target value what we are trying to measure), we are interested in capturing the conversion rate so each user session will consist of a binary variable:

  -  0 - The user did not buy the product 
  -  1 - The user bought the product 

<b> Choosing a sample size </b>

It is important to note that since we won’t test the whole user base (our population), the conversion rates that we’ll get will inevitably be only estimates of the true rates.

The number of people we decide to capture in each group will have an effect on the precision of our estimated conversion rates: the larger the sample size, the more precise our estimates (i.e. the smaller our confidence intervals), the higher the chance to detect a difference in the two groups, if present.

On the other hand, the larger our sample gets, the more expensive (and impractical) our study becomes.

<b><i> So how many people should we have in each group? </i></b>

The sample size we need is estimated through Power Analysis, which depends on a few factors:

  -  Power of the test (1 — β) — This represents the probability of finding a statistical difference between the groups in the test when a difference is actually present. It is the probability that the test correctly rejects the null hypothesis when a specific alternative hypothesis is true. The statistical power ranges from 0 to 1, and as the power of a test increases, the probability β of making a type II error by wrongly failing to reject the null hypothesis decreases. This is usually set at 0.8 by convention.
  
  -  Alpha value (α) — The critical value that was set to 0.05. This is the probability of a type I error.
  
  -  Effect size — How big of a difference we expect there to be between the conversion rates.

Since we want to see a difference of 2 %, we can use 13 % and 15 % to calculate the effect size we expect.

The illustration below shows the power and the significance level of a statistical test that show how the change in the type I error affects the value of the type II error, given the null hypothesis (sampling distribution 1) and the alternative hypothesis (sampling distribution 2).



```python
# importing relevant tools and modules

import numpy as np
import pandas as pd
import scipy.stats as stats
import statsmodels.stats.api as sms
from statsmodels.stats.proportion import proportions_ztest, proportion_confint
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
from math import ceil

%matplotlib inline

# plot styling preferences

plt.style.use('seaborn-whitegrid')
font = {'family' : 'Helvetica',
        'weight' : 'bold',
        'size'   : 14}

mpl.rc('font', **font)
```

```python
# Calculating effect size based on our expected rates

effect_size = sms.proportion_effectsize(0.11, 0.13)
```

```python
# # Calculating sample size needed

required_n = sms.NormalIndPower().solve_power(effect_size, power=0.8, alpha=0.05, ratio=1)
required_n
```

```python
# Rounding up to the next whole number

required_n = ceil(required_n)
required_n
```

We will need at least 4138 observations for each group.

Having set the power parameter to 0.8 in practice means that if there exists an actual difference in conversion rate between our designs, assuming the difference is the one we estimated (11% vs. 13%), we have about 80% chance to detect it as statistically significant in our test with the sample size we calculated.

## 2. Collecting and Preparing the Data

```python
# Loading data

df = pd.read_csv('data/ab_data.csv')
df.head()
```

```python
df.info()
```

```python
# Showing crosstab of the two landing pages with the two groups of people

pd.crosstab(df['group'], df['landing_page'])
```

There are 294478 rows in the DataFrame, each representing a user session, as well as 5 columns :

  -  user_id - The user ID of each session
  -  timestamp - Timestamp for the session
  -  group - Which group the user was assigned to for that session {control, treatment}
  -  landing_page - Which design each user saw on that session {old_page, new_page}
  -  converted - Whether the session ended in a conversion or not (binary, 0=not converted, 1=converted)

From the crosstab we can see that most from the control group have been shown the old page and most from the treatment group the new one.

```python
# Checking how many users have been sampled multiple times

session_counts = df['user_id'].value_counts(ascending=False)
multi_users = session_counts[session_counts > 1].count()

print(f'There are {multi_users} users that appear multiple times in the dataset')
```

There are, in fact, 3894 users that appear more than once. Since the number is low we can remove them from the DataFrame to avoid sampling the same users twice.

```python
# Dropping duplicates

users_to_drop = session_counts[session_counts > 1].index

df = df[~df['user_id'].isin(users_to_drop)]
print(f'The updated dataset now has {df.shape[0]} entries')
```

Now that our DataFrame is nice and clean, we can proceed and sample n= 4138 entries for each of the groups. 

```python
# Random Sampling for control and treatment groups

control_sample = df[df['group'] == 'control'].sample(n=required_n, random_state=42)
treatment_sample = df[df['group'] == 'treatment'].sample(n=required_n, random_state=42)

ab_test = pd.concat([control_sample, treatment_sample], axis=0)
ab_test.reset_index(drop=True, inplace=True)

ab_test
```

```python
ab_test.info()
```

```python
ab_test['group'].value_counts()
```

## 3. Visualizing the Results

```python
# Statistics for samples

conversion_rates = ab_test.groupby('group')['converted']

std_p = lambda x: np.std(x, ddof=0) 
se_p = lambda x: stats.sem(x, ddof=0) 

conversion_rates = conversion_rates.agg([np.mean, std_p, se_p])
conversion_rates.columns = ['conversion_rate', 'std_deviation', 'std_error']
conversion_rates.style.format('{:.3f}')
```

It does look like the two designs performed very similarly, with the new design performing slightly better, approx. 11.3 % vs. 12.7 % conversion rate.

```python
# Plotting the samples

plt.figure(figsize=(8,6))

sns.barplot(x=ab_test['group'], y=ab_test['converted'], ci=False)

plt.ylim(0, 0.17)
plt.title('Conversion rate by group', pad=20)
plt.xlabel('Group', labelpad=15)
plt.ylabel('Converted (proportion)', labelpad=15)
plt.show()
```

The conversion rates for the two groups are very close, but there seems to be a slight increase in the conversion rate to the treatment group. However, to know whether this is by chance, we need to perform a statistical test.

## 4. Testing the Hypothesis

The last step of the analysis is to test the hypothesis. We are analyzing a binary value for which we an perform a Fisher's exact test, but since we have a very large sample we can also approximate the binary samples coming from a normal distribution and use the z-test for calculating the p-value.

```python
# Setting up z-test

control_results = ab_test[ab_test['group'] == 'control']['converted']
treatment_results = ab_test[ab_test['group'] == 'treatment']['converted']

n_con = control_results.count()
n_treat = treatment_results.count()

successes = [control_results.sum(), treatment_results.sum()]
nobs = [n_con, n_treat]
```

```python
# Performing z-test and calculating confidence intervals

z_stat, pval = proportions_ztest(successes, nobs=nobs)

(lower_con, lower_treat), (upper_con, upper_treat) = proportion_confint(successes, nobs=nobs, alpha=0.05)

print(f'z-statistic: {z_stat:.2f}')
print(f'p-value: {pval:.3f}')
print(f'confidence interval 95% for control group: [{lower_con:.3f}, {upper_con:.3f}]')
print(f'confifence interval 95% for treatment group: [{lower_treat:.3f}, {upper_treat:.3f}]')
```

```python
# Alternative Fisher Exact Test

table = np.array([[control_results.sum(), n_con-control_results.sum()], [treatment_results.sum(), n_treat-treatment_results.sum()]])
oddsr, p = stats.fisher_exact(table, alternative='two-sided')
print(f'p-value: {p:.3f}')
```

## 5. Drawing Conclusions

Since both the p-value for the z-test is 0.042 and the p-value for the Fisher's Exact test is 0.046 are below 0.05, we  reject the Null hypothesis Hₒ, which means that the new design did indeed perform (statistically) significantly differently or better than the old design.

Additionally, if we look at the confidence interval for the treatment group ([0.117, 0.138], or 11.7 - 13.8 %) we notice that:

   - It doesn't include the baseline value of 11 % conversion rate.
   - It includes the target value of 13 %.

This means that it is more likely that the true conversion rate of the new design is different to the baseline. This is further proof that the new design has worked and, indeed, is an improvement compared to the old design.
